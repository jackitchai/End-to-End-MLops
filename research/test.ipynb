```python
from sklearn.linear_model import ElasticNet
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python

train_data = pd.read_csv("/Users/kitchaichawmengchai/Downloads/End to End MLops/End-to-End-MLops/artifacts/data_transformation/train.csv")
test_data = pd.read_csv("/Users/kitchaichawmengchai/Downloads/End to End MLops/End-to-End-MLops/artifacts/data_transformation/test.csv")
```

```python
train_data.columns
```

```python
train_x = train_data.drop(["quality"], axis=1)
test_x = test_data.drop(["quality"], axis=1)
train_y = train_data[["quality"]]
test_y = test_data[["quality"]]
```

```python
train_y.columns
```

```python
alpha = 0.9
l1_ratio = 0.8
```

```python
lr = ElasticNet(alpha=alpha, l1_ratio=l1_ratio, random_state=42)
lr.fit(train_x, train_y)
```

```python
pred = lr.predict(test_x)
```

```python
actual = test_y
```

```python
rmse = np.sqrt(mean_squared_error(actual, pred))
mae = mean_absolute_error(actual, pred)
r2 = r2_score(actual, pred)
```

```python
print(f"mae: {mae}")
print(f"r2: {r2}")
print(f"rmse: {rmse}")
#0.4,0.3
# mae: 0.5564786931792307
# r2: 0.21766043946525582
# rmse: 0.6954685784982325
```
